In order to count the measurements for each station in the map-reduce framework we need (for each line):

**Mapper:**
The key will be the station ID and the value is the total number of measurements. We only emit a key-value pair if the measurement type is the type of interest. We could also emit a key-value pair for each single measurement, but this would cause extra work for the reducers.

**Reducer:**
Aggregate all measurement counts.

**Combiner**
We don't need a combiner, but if a mapper encounters multiple lines with same station (this happens since each year gets it own line) we could add a combiner that sums all the measurement counts before sending the data to the reducers.

```python
%%writefile count_measurements.py
#!/usr/bin/python
"""
count the number of measurements of each type
"""
import sys
sys.path.append('/usr/lib/python2.6/dist-packages')
from mrjob.job import MRJob
import re
from sys import stderr

class count_measurements(MRJob):

    def mapper(self, _, line):
        try:
            self.increment_counter('MrJob Counters','mapper',1)
            # Parse the line
            elements=line.split(',')
            station = elements[0]
            measure_type = elements[1]
            year = elements[2]
            measurements = elements[3:]
            no_of_measurements = sum([1 for m in measurements if m != ""])
            
            # Emit the key-value pair
            if measure_type == 'PRCP' and station != 'station': #filter out the unwanted measurements and the header line
                self.increment_counter('MrJob Counters','usefull lines',1)
                yield (station, no_of_measurements)
                
        except Exception, e:
            stderr.write('Error in line:\n'+line)
            stderr.write(e)
            self.increment_counter('MrJob Counters','mapper-error',1)
            yield (('error','mapper', str(e)), 1)
            
    def reducer(self, word, counts):
        try:
            if word[0] == 'error':
                yield(key,sum(data))
                return
            self.increment_counter('MrJob Counters','reducer',1)
            yield (word, sum(counts))
        except Exception, e:
            yield (('error','reducer', str(e)), 1)
        
    def combiner(self, word, counts):
        try:
            self.increment_counter('MrJob Counters','combiner',1)
            yield (word, sum(counts))
        except Exception, e:
            yield (('error','combiner', str(e)), 1)
        
if __name__ == '__main__':
    count_measurements.run()
```

## 2) Efficient spatial data structure for geo data

We have around 9 million years of measurements. To fit a good model of the data, it would be smart to partition the data into smaller areas where the data might have similar patterns and fit a model on each tile. If we blindly split the world into smaller tiles, we might end up having areas with few (or no) data points and areas with many data point, leading to overfitting in some parts of the world and underfitting in other parts. We would rater partition the data into areas of equal number of measurements.

We can do this by dividing the data points recursively using an R-tree ([Wikipedia](http://en.wikipedia.org/wiki/R-tree)).

There are different heristics defining how to split the data. Here we will just take a simple approach and split the data recursively into two parts with alternating latitude and longitude splits.

```python
class rtree_node:
    def __init__(self, dataframe, split_direction='longitude', lable=0, level=0, threshold=10000000, original_data=merged):
        
        # Add labels
        original_data.ix[dataframe.index,'level_%d' % level] = lable
        
        # Split
        group_1, group_2, split_at = self.find_best_split(dataframe, split_direction)
        
        # Recursion base step
        if group_1.sum(axis=0)['measurement_count'] < threshold or group_2.sum(axis=0)['measurement_count'] < threshold:
            self.is_leaf = True
            self.data = dataframe
        # Recursive step
        else:
            self.lable = lable
            self.level = level
            self.is_leaf = False
            self.split_direction = split_direction
            self.split_at = split_at
            self.children = [rtree_node(group_1, self.switch_direction(split_direction), self.lable*2, self.level+1),
                             rtree_node(group_2, self.switch_direction(split_direction), self.lable*2+1, self.level+1)]
    
    # Helper functions
   
    def print_(self, level=0):
        if self.is_leaf:
            print '  '*level, "# of measurements: ", sum([elem for elem in self.data['measurement_count']])
        else:
            self.children[0].print_(level+1)
            print '  '*level, 'split at:', self.split_at
            self.children[1].print_(level+1)

    def switch_direction(self, direction):
        latlon = ['latitude', 'longitude']
        return latlon[latlon.index(direction)-1]
        
    def find_best_split(self, dataframe, split):
        sorted_cumsum = dataframe.sort(split)['measurement_count'].cumsum()
        split_at = dataframe.sum(axis=0)['measurement_count']/2 # half of the sum
        group_1 = sorted_cumsum[sorted_cumsum <= split_at]
        group_2 = sorted_cumsum[sorted_cumsum > split_at]
        return dataframe.loc[group_1.index,:], dataframe.loc[group_2.index,:], split_at
```
